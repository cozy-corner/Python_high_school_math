```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd

dat = pd.read_csv('salary.csv', encoding='SHIFT-JIS')

# データセット
x = dat['年齢']
y = dat['年収']

plt.plot(x, y)
plt.grid(color='0.8')
plt.show()
```

```python
# 差分グラフ
cnt = len(dat)

diff_y = []
for i in range(0, cnt-1):
    diff_y.append(y[i+1] - y[i])
    
plt.plot(x[1:], diff_y)
plt.grid(color='0.8')
plt.show()
```

```python
# 導関数
import numpy as np

x = np.arange(-10, 10, 0.1)
y = x**3 + 3*x**2 + 3*x + 1

plt.plot(x, y)
plt.grid(color='0.8')
plt.show()

y2 = 3*x**2 + 6*x + 3
plt.plot(x, y2)
plt.grid(color='0.8')
plt.show()
```

```python
# 積分

def calc_area(dx):
    h = 0
    area = 0
    
    # 棒の数
    cnt = int(10 / dx)
    for i in range(1, cnt+1):
        h = i * dx
        s = h * dx
        area += s
    return area

calc_area(0.0001)
```

```python
# 曲線の接線
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# 元の関数
x = np.arange(-1, 1, 0.1)
y = 2*x*x + 3

# 接線
a = 4*0.25
b = 3.125 - a * 0.25
y2 = a*x + b

plt.plot(x, y)
plt.plot(x, y2)
plt.grid(color='0.8')
plt.show()
```
