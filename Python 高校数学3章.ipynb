```python
# 連立方程式
from sympy import Symbol, solve

# 式を定義
a = Symbol('a')
b = Symbol('b')

# 両辺が0であること
ex1 = a + b - 1
ex2 = 5*a + b - 3

# タプルの括弧
print(solve((ex1, ex2)))
```

```python
# 基となる線分の傾きと切片
a1 = (5-1)/(6-0)

# 切片
b1 = 1

# 中点
cx = (0+6) / 2
cy = (1+5) / 2

# 直交する線の傾き
a2 = -1 / a1

# 直行する切片
# 座標を元に求める
# y = ax + b 
# b = ax - y
b2 = cy - a2*cx
print(b2)

# 直線の式
x = np.arange(0, 7)
y1 = a1*x + b1
y2 = a2*x + b2

plt.plot(x, y1)
plt.plot(x, y2)

plt.axis('equal')
plt.grid(color='0.8')
plt.show()
```

```python

# 角度
th = np.arange(0, 360)

# 円周上の点Pの座標

# 半径
r = 5

# ラジアン=弧度法 円に対する弧の長さの比率
x = r * np.cos(np.radians(th))
y = r * np.sin(np.radians(th))

plt.plot(x, y)
plt.axis('equal')
plt.grid(color='0.8')
plt.show()          
```

```python

x = [1, 3, 6]
y = [5, 1, 4]

# 辺 AB
a = math.sqrt((x[1] - x[0])**2 + (y[1]-y[0])**2)
print(a)

# 辺 BC
b = math.sqrt((x[2] - x[1])**2 + (y[2]-y[1])**2)
print(b)

# 辺 AC
c = math.sqrt((x[2] - x[0])**2 + (y[2]-y[0])**2)
print(c)

s = (a + b + c) / 2
math.sqrt(s * (s-a) * (s-b) * (s-c))
```

```python
# 三角形の面積

x = [1, 3, 6]
y = [5, 1, 4]

# 辺AB
a = math.sqrt((x[1]-x[0])**2 + (y[1]-y[0])**2)

# 辺BC
b = math.sqrt((x[2]-x[1])**2 + (y[2]-y[1])**2)

# 辺AC
c = math.sqrt((x[2]-x[0])**2 + (y[2]-y[0])**2)

s = (a+b+c) / 2
math.sqrt(s * (s-a) * (s-b) * (s-c))
```
